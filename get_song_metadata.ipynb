```python
import pydub
import pandas as pd
import glob
import os
from pathlib import Path
```

```python
def flatten(d, parent_key='', sep='_'):
    items = []
    for k, v in d.items():
        new_key = parent_key + sep + k if parent_key else k
        if isinstance(v, dict):
            items.extend(flatten(v, new_key, sep=sep).items())
        else:
            items.append((new_key, v))
    return dict(items)
```

```python
df = pd.DataFrame()
```

```python
for song in glob.glob('Jay Chou/*/*.mp3', recursive=True):
    mediainfo = pydub.utils.mediainfo(song)
    df = df.append(flatten(mediainfo), ignore_index=True)
```

```python
titles = []
albums = []
for filename in df.filename:
    fn = Path(filename)
    albums.append(fn.parts[1])
    fn = '.'.join(fn.name.split('.')[-2:])
    fn = fn.split('-')[-1].strip()
    titles.append(fn)
```

```python
df.TAG_title = titles
```

```python
df.TAG_title.head()
```

```python
df.TAG_album = albums
```

```python
df.TAG_album.head()
```

```python
def transform_artist(art):
    try:
        art = art.encode('latin1').decode('gbk')
    except:
        art = '周杰伦'
    return art

df.TAG_artist = df.TAG_artist.apply(transform_artist)
```

```python
df.drop('TAG_comment', axis=1, inplace=True)
```
